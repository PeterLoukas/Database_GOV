# Δημιουργία πινάκων!

```python
%%sql

CREATE TABLE GOV_PLATFORM (Timestamp time(0), IP_Address varchar(15) not null, cookies varchar(100), URL_Address varchar(50) not null, Name varchar(20), Device_Connection_Elements varchar (40), PRIMARY KEY (IP_Address, URL_Address));
CREATE TABLE CITIZEN(c_first_name varchar(15), last_name varchar(20), c_postal_code int, father_name varchar(15), mother_name varchar(15), c_email varchar(50), c_key_number varchar(22) not null, c_phone varchar(25), ID_number varchar(8) not null, c_address varchar(25), place_of_birth varchar(15), c_taxpayer_ID int not null, social_security_number int, year_of_birth date, IP_Address varchar(15), URL_Address varchar(50), suggestions varchar(50), questions varchar(50), PRIMARY KEY (c_Taxpayer_ID, c_key_number, ID_number),CHECK (year_of_birth >= 1900), FOREIGN KEY (IP_Address, URL_Address) REFERENCES GOV_PLATFORM(IP_Address, URL_Address));
CREATE TABLE BUSINESS (b_name varchar(25), owner_name varchar(30), activity varchar(20),b_Taxpayer_ID int not null, b_postal_code int, activity_number_code varchar(20) , b_email varchar(30), b_phone varchar(25), b_address varchar(25), b_key_number varchar(22) not null, IP_Address varchar(15), URL_Address varchar(50), suggestions varchar(60) not null, questions varchar(60) not null, PRIMARY KEY (b_key_number, b_Taxpayer_ID), FOREIGN KEY (IP_Address, URL_Address)REFERENCES GOV_PLATFORM(IP_Address, URL_Address));
CREATE TABLE SERVICES (Ministry varchar(25), Independent_Authority varchar(30), Regulators_Organizations varchar(40), Districts varchar(25), Level_1_Category varchar(35), Level_2 _Life Event varchar(35), Level_3_Service_transaction varchar(25), IP_Address varchar(15), URL_Address varchar(50), Edifying_Gate varchar(40), Scripture_Digital_Reformation varchar(45), Digital_Gate_ERMIS varchar(60), Quick_Search varchar(60), Compound_Search varchar(60), Topical_Search varchar(60), FOREIGN KEY (IP_Address, URL_Address) REFERENCES GOV_PLATFORM(IP_Address, URL_Address));
CREATE TABLE BUSINESS_E_FILE(Data_e_books varchar(5000), b_key_number int not null, registers varchar(22), PRIMARY KEY(b_key_number));
CREATE TABLE C_PHONE(c_phone varchar(25),c_key_number varchar(22) not null,ID_number varchar(8) not null,PRIMARY KEY(c_phone,c_key_number,ID_number));
CREATE TABLE B_PHONE(b_phone varchar(25),b_key_number varchar(22) not null,b_Taxpayer_id int not null,PRIMARY KEY(b_phone,b_key_number,b_Taxpayer_id));
CREATE TABLE CONNECT_C(c_key_number varchar(22) not null,c_Taxpayer_ID int not null,ID_number varchar(8) not null,IP_Address varchar(15),URL_Address int,c_Authentication varchar(50) not null,c_web_banking int not null,c_Taxis_Net int not null,PRIMARY KEY(c_key_number,c_Taxpayer_id,ID_number,c_Authentication,c_web_banking,c_Taxis_Net),FOREIGN KEY(c_key_number,c_Taxpayer_ID,ID_number)REFERENCES CITIZEN(c_key_number,c_Taxpayer_ID,ID_number),FOREIGN KEY(IP_Address,URL_Address)REFERENCES GOV_PLATFORM(IP_Address,URL_Address),FOREIGN KEY(c_Authentication,c_web_banking,c_Taxis_net)REFERENCES CONNECT_C(c_Authentication,c_web_banking,c_Taxis_net));
CREATE TABLE CONNECT_B (b_Key_Number varchar(22) not null, b_Taxpayer_ID int not null, IP_Address varchar(15), URL_Address varchar(50), b_Authentication varchar(50) not null, b_web_banking int not null, b_Taxis_Net int not null, PRIMARY KEY (b_Key_Number, b_Taxpayer_ID, b_Authentication, b_web_banking, b_Taxis_Net), FOREIGN KEY (b_Key_Number, b_Taxpayer_ID)REFERENCES BUSSINESS(b_Key_Number, b_Taxpayer_ID),FOREIGN KEY(IP_Address, URL_Address)REFERENCES GOV_PLATFORM(IP_Address, URL_Address),FOREIGN KEY(b_Authentication, b_web_banking, b_Taxis_Net)REFERENCES CONNECT_B(b_Authentication, b_web_banking, b_Taxis_Net));
CREATE TABLE SUBMIT (c_key_number varchar(22) not null, ID_number varchar(8) not null, c_Taxpayer_ID int not null, b_Taxpayer_ID int not null, b_key_number varchar(22) not null, IP_Address varchar(15) not null, URL_Address varchar(50) , suggestions varchar(50), questions varchar(50), PRIMARY KEY (c_key_number, ID_number, c_Taxpayer_ID, b_Taxpayer_ID, b_key_number, suggestions, questions), FOREIGN KEY(IP_Address, URL_Address)REFERENCES GOV_PLATFORM(IP_Address, URL_Address),FOREIGN KEY(c_key_number, ID_number, c_Taxpayer_ID)REFERENCES CITIZEN(c_key_number, ID_number, c_Taxpayer_ID),FOREIGN KEY(b_Taxpayer_ID, b_key_number)REFERENCES BUSINESS(b_Taxpayer_ID, b_key_number), FOREIGN KEY(suggestions, questions)REFERENCES SUBMIT (suggestions, questions));
```

# Ερωτήματα βάσης!

Να εμφανιστούν οι Συσκευές που Συνδέθηκαν οι χρήστες μέσω των οποίων η σύνδεση στην πλατφόρμα gov.gr διήρκησε περισσότερο από 90 και λιγότερο από 200 δευτερόλεπτα, ταξινομημένες σε αύξουσα διάταξη.

```python
%%sql

SELECT Device_Connection_Elements, Timestamp as "Seconds"
FROM GOV_PLATFORM 
WHERE Timestamp BETWEEN 90.0 and 200.00 ORDER BY Timestamp ASC; 
```

Τα ΑMKA των Πολιτών που οι μητέρες τους έχουν το ίδιο όνομα, ταξινομημένα βάσει την χρονολογία γέννησης τους. 

```python
%%sql

SELECT social_security_number,  mother_name, year_of_birth
FROM CITIZEN 
WHERE mother_name in
(
    SELECT mother_name
    FROM CITIZEN 
    GROUP BY mother_name 
    HAVING count(*)>1
) 
ORDER BY year_of_birth;
```

Να εμφανιστεί ελάχιστος ο Αριθμός Δραστηριότητας των εταιριών εκείνων οι οποίες έχουν αριθμό οδού '2' και χρησιμοποιούν gmail.

```python
%%sql

SELECT DISTINCT min(activity_number_code), b_name, b_address, b_email
FROM BUSINESS
WHERE  b_email LIKE "%gmail%" AND b_address LIKE ('%2');
```

Ποια τα Ονοματεπώνυμα τον Πολιτών οι οποίοι γεννήθηκαν στην Καβάλα, ο αριθμός τηλεφώνου ξεκινάει με '695' ή δεν αρχίζει ο αριθμός ταυτότητας τους από 'AM';

```python
%%sql

SELECT c_first_name, last_name, place_of_birth, c_phone, ID_number
FROM CITIZEN
WHERE (ID_number NOT LIKE "%AM%") OR (place_of_birth = 'KAVALA') AND (CITIZEN.c_key_number, CITIZEN.ID_number) in
    (
        SELECT C_PHONE.c_key_number, C_PHONE.ID_number
        FROM C_PHONE
        WHERE c_phone LIKE "%695%" 
    );
    
```

Ποιος είναι ο Μέσος Χρόνος Σύνδεσης στην πλατφόρμα gov.gr;

```python
%%sql

SELECT AVG(Timestamp)
FROM GOV_PLATFORM
```

Ποιοι είναι οι Κλειδάριθμοι των Επιχειρήσεων οι οποίες δραστηριοποιούνται στον παραγωγικό-κατασκευαστικό τομέα;

```python
%%sql

SELECT b_key_number, activity
FROM BUSINESS
WHERE activity_number_code LIKE ('2%');
```

Πόσες Φορές αναζητήκε το "Υπουργείο Μετανάστευσης και Ασύλου" μέσω σύνθετης αναζήτησης;

```python
%%sql

SELECT count(Compound_Search), Ministry, Compound_Search
FROM SERVICES
WHERE (Ministry = 'Ministry of Immigration and Asylum');
```

Ποια είναι η Δραστηριότητα των εταιριών οι οποίες δεν υπέβαλλαν κάποιο ερώτημα ή οι ιδιοκτήτες τους έχουν το επίθετο  Καπετάν;

```python
%%sql

SELECT BUSINESS.activity, questions, owner_name
FROM BUSINESS
WHERE (owner_name LIKE "%KAPETAN%") or (BUSINESS.b_key_number, BUSINESS.b_Taxpayer_ID) in
(
    SELECT CONNECT_B.b_key_number, CONNECT_B.b_Taxpayer_ID
    FROM CONNECT_B
    WHERE questions = '-'    
);
```

Ποια είναι τα Ονομάτα των Επιχειρήσεων όπου ο ταχτυδρομικός τους κώδικας είναι 67100;

```python
%%sql

SELECT BUSINESS.b_name, b_postal_code
FROM BUSINESS
WHERE (BUSINESS.b_key_number, BUSINESS.b_Taxpayer_ID) in
(
    SELECT CONNECT_B.b_key_number, CONNECT_B.b_Taxpayer_ID
    FROM CONNECT_B
    WHERE (CONNECT_B.b_key_number, CONNECT_B.b_Taxpayer_ID) in
        (
            SELECT SUBMIT.b_key_number, SUBMIT.b_Taxpayer_ID
            FROM SUBMIT
            WHERE b_postal_code = '67100'
        )
);
```
